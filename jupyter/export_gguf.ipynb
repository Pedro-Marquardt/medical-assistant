```python
caminho_lora_drive = "/content/drive/MyDrive/mistral_lora_postech"

# 3. Carrega o modelo base já "vestido" com os seus pesos do hospital
print("\nCarregando o modelo e os pesos LoRA...")
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = caminho_lora_drive,
    max_seq_length = 2048,
    dtype = None,
    load_in_4bit = True,
)
```

```python
nome_base_arquivo = "mistral_medico_postech"

model.save_pretrained_gguf(
    nome_base_arquivo,
    tokenizer,
    quantization_method = "q4_k_m"
)

# 5. Move o arquivo final pronto para o seu Drive para você baixar no Mac
arquivo_gerado = f"{nome_base_arquivo}-unsloth.Q4_K_M.gguf"
destino_drive = f"/content/drive/MyDrive/{arquivo_gerado}"
```
