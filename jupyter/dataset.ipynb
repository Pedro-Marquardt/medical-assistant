```python
# BLOCK 3: MERGE AND FORMAT (ALPACA STANDARD)

def format_to_alpaca(row):

    alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
Você é um assistente médico especializado do Hospital PosTech. Forneça respostas técnicas, seguras e baseadas nos protocolos internos.

### Input:
{}

### Response:
{}"""
    

    EOS_TOKEN = "</s>" 
    
    formatted_text = alpaca_prompt.format(row['question'], row['answer']) + EOS_TOKEN
    return formatted_text

print("Merging MedQuAD and Synthetic datasets...")
df_complete = pd.concat([df_medquad, synthetic_data], ignore_index=True)

print("Applying Alpaca formatting...")
df_complete['text'] = df_complete.apply(format_to_alpaca, axis=1)

print("\n--- Alpaca Format Preview ---")
print(df_complete['text'].iloc[-1]) # Shows the last synthetic row added
```
